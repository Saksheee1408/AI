```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, classification_report
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Conv1D, MaxPooling1D, UpSampling1D, Flatten, Reshape, Conv1DTranspose, Lambda
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import os

# 1. Load the dataset
def load_data(normal_path, anomaly_path):
    normal_df = pd.read_csv('/content/ptbdb_normal.csv').iloc[:, :-1]  # Exclude last column
    anomaly_df = pd.read_csv('/content/ptbdb_abnormal.csv').iloc[:, :-1]

    # Add labels: 0 for normal, 1 for anomaly
    normal_df['target'] = 0
    anomaly_df['target'] = 1

    # Combine datasets
    all_df = pd.concat([normal_df, anomaly_df], axis=0).reset_index(drop=True)

    # Extract features and labels
    X = all_df.iloc[:, :-1].values
    y = all_df['target'].values

    # Normalize the data
    X = (X - np.min(X)) / (np.max(X) - np.min(X))

    # Reshape for CNN input [samples, timesteps, features]
    X = X.reshape(X.shape[0], X.shape[1], 1)

    return X, y

# 2. Split data into train, validation, test sets
def split_data(X, y, test_size=0.2, val_size=0.2):
    # First split: training + validation vs test
    X_train_val, X_test, y_train_val, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42, stratify=y
    )

    # Second split: training vs validation
    X_train, X_val, y_train, y_val = train_test_split(
        X_train_val, y_train_val, test_size=val_size, random_state=42, stratify=y_train_val
    )

    return X_train, X_val, X_test, y_train, y_val, y_test

# 3. Build an autoencoder model with transposed convolution for better reconstruction
def build_autoencoder(input_shape):
    # Input layer
    input_layer = Input(shape=input_shape)

    # Encoder
    x = Conv1D(32, 5, activation='relu', padding='same')(input_layer)
    x = MaxPooling1D(2, padding='same')(x)
    x = Conv1D(64, 5, activation='relu', padding='same')(x)
    x = MaxPooling1D(2, padding='same')(x)
    x = Conv1D(128, 5, activation='relu', padding='same')(x)
    encoded = MaxPooling1D(2, padding='same')(x)

    # Decoder with transposed convolution (better than upsampling)
    x = Conv1DTranspose(128, 5, strides=2, activation='relu', padding='same')(encoded)
    x = Conv1DTranspose(64, 5, strides=2, activation='relu', padding='same')(x)
    x = Conv1DTranspose(32, 5, strides=2, activation='relu', padding='same')(x)

    # Add a Lambda layer to ensure the output shape matches input shape
    x = Lambda(lambda y: y[:, :input_shape[0], :])(x)

    decoded = Conv1D(1, 5, activation='sigmoid', padding='same')(x)

    # Autoencoder model
    autoencoder = Model(input_layer, decoded)
    # Use tf.keras.losses.MeanSquaredError() instead of just 'mse' string
    autoencoder.compile(optimizer='adam', loss=tf.keras.losses.MeanSquaredError())

    return autoencoder

# 4. Train the autoencoder on normal data only
def train_autoencoder(autoencoder, X_train, X_val, y_train, y_val, epochs=50, batch_size=32):
    # Get only normal samples for training
    X_train_normal = X_train[y_train == 0]
    X_val_normal = X_val[y_val == 0]

    # Early stopping to prevent overfitting
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True
    )

    # Model checkpoint to save best model
    checkpoint = ModelCheckpoint(
        'best_autoencoder.h5',
        monitor='val_loss',
        save_best_only=True,
        mode='min'
    )

    # Train the model
    history = autoencoder.fit(
        X_train_normal, X_train_normal,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_val_normal, X_val_normal),
        callbacks=[early_stopping, checkpoint],
        verbose=1
    )

    return autoencoder, history

# 5. Calculate reconstruction error (MSE) for each sample
def calculate_reconstruction_error(autoencoder, X):
    X_pred = autoencoder.predict(X)
    mse = np.mean(np.square(X - X_pred), axis=(1, 2))
    return mse

# 6. Find optimal threshold for anomaly detection
def find_optimal_threshold(autoencoder, X_val, y_val):
    # Calculate reconstruction error for validation set
    val_error = calculate_reconstruction_error(autoencoder, X_val)

    # Try different thresholds and find the one with the best F1 score
    thresholds = np.linspace(np.min(val_error), np.max(val_error), 100)
    best_f1 = 0
    best_threshold = thresholds[0]  # Initialize with first threshold instead of None

    for threshold in thresholds:
        y_pred = (val_error > threshold).astype(int)
        f1 = f1_score(y_val, y_pred)

        if f1 > best_f1:
            best_f1 = f1
            best_threshold = threshold

    # Add a safeguard in case no good threshold was found
    if best_f1 == 0:
        print("Warning: Could not find a threshold with F1 score > 0. Using mean error as threshold.")
        best_threshold = np.mean(val_error)

    return best_threshold

# 7. Evaluate the model on test data
def evaluate_model(autoencoder, X_test, y_test, threshold):
    # Calculate reconstruction error for test set
    test_error = calculate_reconstruction_error(autoencoder, X_test)

    # Predict anomalies based on threshold
    y_pred = (test_error > threshold).astype(int)

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Create confusion matrix
    cm = confusion_matrix(y_test, y_pred)

    # Print metrics
    print(f"Accuracy: {accuracy*100:.2f}%")
    print(f"Precision: {precision*100:.2f}%")
    print(f"Recall: {recall*100:.2f}%")
    print(f"F1 Score: {f1*100:.2f}%")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    return accuracy, precision, recall, f1

# 8. Visualize results
def plot_results(autoencoder, X_test, y_test, threshold):
    # Get a random sample of normal and anomaly ECGs from test set
    test_error = calculate_reconstruction_error(autoencoder, X_test)
    normal_indices = np.where(y_test == 0)[0]
    anomaly_indices = np.where(y_test == 1)[0]

    # Select 3 random samples from each class
    normal_sample = np.random.choice(normal_indices, 3, replace=False)
    anomaly_sample = np.random.choice(anomaly_indices, 3, replace=False)

    # Plot the reconstruction of normal ECGs
    plt.figure(figsize=(12, 8))
    for i, idx in enumerate(normal_sample):
        # Original signal
        plt.subplot(3, 4, i*4+1)
        plt.plot(X_test[idx].reshape(-1))
        plt.title(f"Normal Original {i+1}")
        plt.grid(True)

        # Reconstructed signal
        plt.subplot(3, 4, i*4+2)
        plt.plot(autoencoder.predict(X_test[idx].reshape(1, -1, 1)).reshape(-1))
        plt.title(f"Normal Reconstructed {i+1}")
        plt.grid(True)

        # Error
        plt.subplot(3, 4, i*4+3)
        plt.plot(np.abs(X_test[idx].reshape(-1) - autoencoder.predict(X_test[idx].reshape(1, -1, 1)).reshape(-1)))
        plt.title(f"Error: {test_error[idx]:.4f}")
        plt.grid(True)

        # Decision
        plt.subplot(3, 4, i*4+4)
        plt.text(0.5, 0.5, f"Threshold: {threshold:.4f}\nPrediction: {'Anomaly' if test_error[idx] > threshold else 'Normal'}",
                 ha='center', va='center', fontsize=10)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

    # Plot the reconstruction of anomaly ECGs
    plt.figure(figsize=(12, 8))
    for i, idx in enumerate(anomaly_sample):
        # Original signal
        plt.subplot(3, 4, i*4+1)
        plt.plot(X_test[idx].reshape(-1))
        plt.title(f"Anomaly Original {i+1}")
        plt.grid(True)

        # Reconstructed signal
        plt.subplot(3, 4, i*4+2)
        plt.plot(autoencoder.predict(X_test[idx].reshape(1, -1, 1)).reshape(-1))
        plt.title(f"Anomaly Reconstructed {i+1}")
        plt.grid(True)

        # Error
        plt.subplot(3, 4, i*4+3)
        plt.plot(np.abs(X_test[idx].reshape(-1) - autoencoder.predict(X_test[idx].reshape(1, -1, 1)).reshape(-1)))
        plt.title(f"Error: {test_error[idx]:.4f}")
        plt.grid(True)

        # Decision
        plt.subplot(3, 4, i*4+4)
        plt.text(0.5, 0.5, f"Threshold: {threshold:.4f}\nPrediction: {'Anomaly' if test_error[idx] > threshold else 'Normal'}",
                 ha='center', va='center', fontsize=10)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# 9. Function to predict on new data
def predict_anomaly(autoencoder, ecg_data, threshold):
    """
    Predict if an ECG signal is normal or anomalous

    Parameters:
    - autoencoder: Trained autoencoder model
    - ecg_data: ECG signal of shape (187,) or (1, 187, 1)
    - threshold: Threshold for anomaly detection

    Returns:
    - prediction: 'Normal' or 'Anomaly'
    - confidence: Confidence score based on reconstruction error
    - error: Reconstruction error value
    """
    # Ensure data is in the right shape
    if len(ecg_data.shape) == 1:
        ecg_data = ecg_data.reshape(1, -1, 1)

    # Calculate reconstruction error
    reconstructed = autoencoder.predict(ecg_data)
    error = np.mean(np.square(ecg_data - reconstructed))

    # Make prediction
    if error > threshold:
        prediction = 'Anomaly'
    else:
        prediction = 'Normal'

    # Calculate confidence score (higher error = higher confidence for anomaly)
    if prediction == 'Anomaly':
        confidence = min(100, (error / threshold) * 100 - 100)
    else:
        confidence = min(100, (1 - (error / threshold)) * 100)

    return prediction, confidence, error

# 10. Function to process uploaded image of ECG
def process_ecg_image(image_path):
    """
    Process an ECG image to extract the signal

    Parameters:
    - image_path: Path to the ECG image

    Returns:
    - extracted_signal: Extracted ECG signal as numpy array
    """
    try:
        import cv2
        from scipy import signal

        # Read the image
        img = cv2.imread(image_path)
        if img is None:
            raise ValueError(f"Could not read image from {image_path}")

        # Convert to grayscale
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Apply thresholding to get binary image
        _, binary = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV)

        # Find contours
        contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Find the largest contour (likely the ECG line)
        largest_contour = max(contours, key=cv2.contourArea)

        # Extract points from the contour
        points = largest_contour.reshape(-1, 2)

        # Sort points by x-coordinate
        points = points[points[:, 0].argsort()]

        # Extract y-coordinates
        y_values = points[:, 1]

        # Normalize y-values (invert as image coordinates are from top to bottom)
        y_values = np.max(y_values) - y_values

        # Resample to 308 points (match our model input)
        if len(y_values) > 308:
            # Downsample using scipy's signal.resample
            extracted_signal = signal.resample(y_values, 308)
        else:
            # If fewer points, use interpolation
            x_old = np.linspace(0, 1, len(y_values))
            x_new = np.linspace(0, 1, 308)
            extracted_signal = np.interp(x_new, x_old, y_values)

        # Normalize to [0, 1]
        extracted_signal = (extracted_signal - np.min(extracted_signal)) / (np.max(extracted_signal) - np.min(extracted_signal))

        # Reshape for model input
        extracted_signal = extracted_signal.reshape(1, 308, 1)

        return extracted_signal

    except Exception as e:
        print(f"Error extracting ECG signal from image: {e}")
        print("Returning a simulated ECG signal instead.")
        # Return simulated signal as fallback
        return np.random.rand(1, 308, 1)
def load_model_safely(model_path):
    """
    Load the model with appropriate error handling
    """
    try:
        # Try the standard way first
        return tf.keras.models.load_model(model_path)
    except TypeError as e:
        print(f"Error loading model: {e}")
        print("Attempting to load with custom objects...")

        # Try with custom objects
        return tf.keras.models.load_model(
            model_path,
            custom_objects={
                'mse': tf.keras.losses.MeanSquaredError()
            }
        )
    except Exception as e:
        print(f"Failed to load model: {e}")
        print("Please retrain the model")
        return None
# 11. Process uploaded ECG image and predict anomaly
def process_uploaded_ecg(image_path, model_path, threshold):
    """
    Process an uploaded image of an ECG and predict if it's normal or anomalous

    Parameters:
    - image_path: Path to the ECG image
    - model_path: Path to the saved model
    - threshold: Threshold for anomaly detection

    Returns:
    - prediction: 'Normal' or 'Anomaly'
    - confidence: Confidence score
    - error: Reconstruction error
    """
    # Load the model with custom loader
    autoencoder = load_model_safely(model_path)

    if autoencoder is None:
        print("Could not load model. Please retrain the model.")
        return "Unknown", 0.0, 0.0

    # Extract ECG signal from image
    extracted_ecg = process_ecg_image(image_path)

    # Make prediction
    prediction, confidence, error = predict_anomaly(autoencoder, extracted_ecg, threshold)

    # Display results
    plt.figure(figsize=(12, 4))

    # Original "extracted" signal
    plt.subplot(1, 3, 1)
    plt.plot(extracted_ecg.reshape(-1))
    plt.title("Extracted ECG Signal")
    plt.grid(True)

    # Reconstructed signal
    plt.subplot(1, 3, 2)
    plt.plot(autoencoder.predict(extracted_ecg).reshape(-1))
    plt.title("Reconstructed Signal")
    plt.grid(True)

    # Result
    plt.subplot(1, 3, 3)
    plt.text(0.5, 0.5, f"Prediction: {prediction}\nConfidence: {confidence:.2f}%\nError: {error:.4f}\nThreshold: {threshold:.4f}",
             ha='center', va='center', fontsize=12)
    plt.axis('off')

    plt.tight_layout()
    plt.show()

    return prediction, confidence, error

# 12. NEW: Google Colab file upload integration
def upload_and_check_ecg():
    """
    Upload an ECG image through Google Colab interface and analyze it
    """
    try:
        from google.colab import files

        print("Please upload an ECG image file...")
        uploaded = files.upload()

        if not uploaded:
            print("No file was uploaded.")
            return

        # Get the filename
        image_path = list(uploaded.keys())[0]

        # Load the model
        model_path = 'ecg_anomaly_detector.h5'
        if not os.path.exists(model_path):
            print(f"Model file {model_path} not found. Please train the model first.")
            return

        # Load the threshold
        threshold_path = 'optimal_threshold.npy'
        if os.path.exists(threshold_path):
            threshold = float(np.load(threshold_path))
            if np.isnan(threshold):
                # If threshold is NaN, use default
                threshold = 0.01
                print(f"Loaded threshold is NaN. Using default threshold: {threshold:.6f}")
            else:
                print(f"Loaded threshold value: {threshold:.6f}")
        else:
            # Use a default threshold if not found
            threshold = 0.01
            print(f"Threshold file not found. Using default threshold: {threshold:.6f}")

        # Process the uploaded ECG image
        print(f"Processing ECG image: {image_path}")
        prediction, confidence, error = process_uploaded_ecg(image_path, model_path, threshold)

        print(f"Prediction: {prediction}")
        print(f"Confidence: {confidence:.2f}%")
        print(f"Error: {error:.6f}")
        print(f"Threshold: {threshold:.6f}")

        return prediction, confidence, error

    except ImportError:
        print("Google Colab is not available in this environment.")
        print("Please use the check_ecg_image function instead.")
        return None

# 13. NEW: Regular file path function for non-Colab environments
def check_ecg_image(image_path):
    """
    Analyze an ECG image from a file path

    Parameters:
    - image_path: Path to the ECG image file

    Returns:
    - prediction: 'Normal' or 'Anomaly'
    - confidence: Confidence score
    - error: Reconstruction error
    """
    # Check if the image file exists
    if not os.path.exists(image_path):
        print(f"Image file {image_path} not found.")
        return None

    # Load the model
    model_path = 'ecg_anomaly_detector.h5'
    if not os.path.exists(model_path):
        print(f"Model file {model_path} not found. Please train the model first.")
        return None

    # Load the threshold
    threshold_path = 'optimal_threshold.npy'
    if os.path.exists(threshold_path):
        threshold = float(np.load(threshold_path))
        if np.isnan(threshold):
            # If threshold is NaN, use default
            threshold = 0.01
            print(f"Loaded threshold is NaN. Using default threshold: {threshold:.6f}")
        else:
            print(f"Loaded threshold value: {threshold:.6f}")
    else:
        # Use a default threshold if not found
        threshold = 0.01
        print(f"Threshold file not found. Using default threshold: {threshold:.6f}")

    # Process the ECG image
    print(f"Processing ECG image: {image_path}")
    prediction, confidence, error = process_uploaded_ecg(image_path, model_path, threshold)

    print(f"Prediction: {prediction}")
    print(f"Confidence: {confidence:.2f}%")
    print(f"Error: {error:.6f}")
    print(f"Threshold: {threshold:.6f}")

    return prediction, confidence, error

# Main execution flow
def main():
    # File paths
    normal_path = "ptbdb_normal.csv"
    anomaly_path = "ptbdb_abnormal.csv"

    # Check if files exist
    if not (os.path.exists(normal_path) and os.path.exists(anomaly_path)):
        print(f"Please ensure the dataset files exist in the current directory:")
        print(f"- {normal_path}")
        print(f"- {anomaly_path}")
        return

    # 1. Load and preprocess data
    print("Loading and preprocessing data...")
    X, y = load_data(normal_path, anomaly_path)

    # 2. Split data
    print("Splitting data into train, validation, and test sets...")
    X_train, X_val, X_test, y_train, y_val, y_test = split_data(X, y)

    # 3. Build and train model
    print("Building and training autoencoder model...")
    input_shape = (X_train.shape[1], 1)  # (187, 1)
    autoencoder = build_autoencoder(input_shape)

    # Print model summary
    autoencoder.summary()

    # 4. Train the model
    print("Training the model on normal data only...")
    autoencoder, history = train_autoencoder(autoencoder, X_train, X_val, y_train, y_val, epochs=50)

    # 5. Plot training history
    plt.figure(figsize=(10, 4))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Autoencoder Training History')
    plt.xlabel('Epoch')
    plt.ylabel('Mean Squared Error')
    plt.legend()
    plt.grid(True)
    plt.show()

    # 6. Find optimal threshold
    print("Finding optimal threshold for anomaly detection...")
    threshold = find_optimal_threshold(autoencoder, X_val, y_val)
    print(f"Optimal threshold: {threshold:.6f}")

    # 7. Evaluate model on test data
    print("Evaluating model on test data...")
    accuracy, precision, recall, f1 = evaluate_model(autoencoder, X_test, y_test, threshold)

    # 8. Visualize results
    print("Visualizing reconstruction results...")
    plot_results(autoencoder, X_test, y_test, threshold)

    # 9. Save the model and threshold
    autoencoder.save('ecg_anomaly_detector.h5')
    np.save('optimal_threshold.npy', threshold)
    print("Model saved as 'ecg_anomaly_detector.h5'")
    print(f"Threshold saved as 'optimal_threshold.npy' with value {threshold:.6f}")

    # 10. Provide instructions for using the upload feature
    print("\nModel is ready for anomaly detection!")
    print("There are two ways to analyze new ECG images:")
    print("1. For Google Colab: Run the upload_and_check_ecg() function")
    print("2. For other environments: Use check_ecg_image('/path/to/ecg_image.jpg')")

# Example of how to use the upload features after training
if __name__ == "__main__":
    # First train the model
    main()

    # Then you can use one of these methods to analyze new ECG images

    # Option 1: For Google Colab - uncomment the line below
    upload_and_check_ecg()

    # Option 2: For local or other environments - uncomment and provide image path
    # check_ecg_image('/path/to/your/ecg_image.jpg')
```
